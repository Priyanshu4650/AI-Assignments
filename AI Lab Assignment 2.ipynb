(b) Let the number of unique words be $n$. Construct an $n \times n$ matrix $M$ where the $i$th row corresponds to the $i$th word, and the $(i, j)$th entry stands for the frequency of occurrence of $j$th word after the $i$th word. Write a function, which accepts a given word, and returns the frequencies of occurrences of the next words (i.e., the row corresponding to the given word).

```python
#Text Generation(b)

#Importing re library
import re

#Importing pandas library as pd
import pandas as pd

import numpy as np

#Created a function to generate the output as asked in the question
def get_frequencies(filename, word):
    
    #Opening file and reading it in lower case
    with open(filename, encoding = 'utf8') as f:
        text = f.read().lower()
        
    # Remove punctuation from the text
    text = re.sub(r'[^\w\s]', '', text)
    
    # Split the text into words
    words = text.split()
    
    # Create a pandas DataFrame from the words
    words_df = pd.DataFrame({'Word': words})
    
    # Create a list of the unique words
    unique_words = words_df['Word'].unique().tolist()
    
    # Create an empty n x n matrix where n is the number of unique words
    n = len(unique_words)
    
    #Creating a 2D list to store the frequencies as asked in the question
    M = np.zeros((n, n))
    
    # Iterate through the words and update the frequency matrix
    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]
        
        current_word_index = unique_words.index(current_word)
        next_word_index = unique_words.index(next_word)
        
        M[current_word_index][next_word_index] += 1
    
    # Find the index of the given word in the list of unique words
    word_index = unique_words.index(word)
    
    # Return the row corresponding to the given word
    return M[word_index, :]

word = input('Enter word to get the frequency :')
frequencies = get_frequencies("speeches.txt", word)

# Print the frequencies of occurrences of the next words
print(frequencies)
```

```python
#Text Generation(c)

#Importing random library
import random

#Importing re library
import re

#Importing numpy library as np
import numpy as np

#Importing pandas library as pd
import pandas as pd

#Creating function to produce random text of 5000 words
def produce_random_text(filename, length):
    
    #Reading the data of the file in lower case
    with open(filename, 'r') as f:
        text = f.read().lower()
        
    # Remove punctuation from the text
    text = re.sub(r'[^\w\s]', '', text)
    
    # Split the text into words
    words = text.split()
    
    # Create a pandas DataFrame from the words
    words_df = pd.DataFrame({'Word': words})
    
    # Create a list of the unique words
    unique_words = words_df['Word'].unique().tolist()
    
    # Create an empty n x n matrix where n is the number of unique words
    n = len(unique_words)
    
    #Creating a 2D list to store the frequencies as asked in the (b) part
    M = np.zeros((n, n))
    
    # Iterate through the words and update the frequency matrix
    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]
        
        current_word_index = unique_words.index(current_word)
        next_word_index = unique_words.index(next_word)
        
        M[current_word_index][next_word_index] += 1
        
    # Normalize the matrix to obtain probabilities
    M = M / np.sum(M, axis=1, keepdims=True)
    
    # Choose a random first word from the list of unique words
    first_word = random.choice(unique_words)
    first_word_index = unique_words.index(first_word)
    
    # Initialize the random text with the first word
    random_text = [first_word]
    
    # Generate the next words based on the probabilities in the frequency matrix
    current_word_index = first_word_index
    for i in range(length - 1):
        next_word_probs = M[current_word_index, :]
        next_word_index = np.random.choice(np.arange(n), p=next_word_probs)
        next_word = unique_words[next_word_index]
        
        random_text.append(next_word)
        current_word_index = next_word_index
    
    return " ".join(random_text)
 
length = 5000
random_text = produce_random_text("speeches.txt", length)

# Print the random text
print(random_text)
```

(b) Implement an agent which at each time instant $t$ observes its position $o_t =(x_t, y_t)$  (but not the dirt information in this location) and performs a random action $a_t \in \{up,down,right,left, pickDirt\}$. On hitting a wall or picking dirt from the current location $(x_t, y_t)$, the agent stays in the same position $(x_t, y_t)$, otherwise, its position changes according to the action. Every action is associated with a reward (or penalty) defined as follows.

   
$$r_t = R((x_t, y_t), a_t) = 
     \begin{cases}
       -1, &\text{if the agent tries to pick dirt and } (x_t, y_t) \text{ is a clean cell}\\
      -10, &\text{if the agent hits a wall}
       \\ 
       d, &\text{if } (x_t, y_t) \text{ has } d \text{ units of dirt and the agent picks it}\\
       0, &\text{otherwise }
     \end{cases}
$$



```python
#Robot Cleaning Agent (b)

#Importing random and numpy library functions
import random
import numpy as np

#Creating a list of actions to take
actions = ['up', 'down', 'left', 'right', 'pickDirt']

#Function to add 10 dirt particles on the floor
def add_dirt(room):
    
    #Running loop 10 times to add 10 dirt partcles
    for i in range(10):
        
        #Randomly selecting the coordinates to add dirt at that place
        x = random.randint(0, m-1)
        y = random.randint(0, n-1)
        
        #Incrementing the value of dirt at that place
        room[x][y] += 1

#Function to implement robots action
def agents_Action(room, t):
    
    #Selecting random coordinates to start the task
    x = random.randint(0, m-1)
    y = random.randint(0, n-1)
    
    #List to store the rewards at every time t
    reward = []
    
    #Temporary variable to store reward
    sum=0
    
    #Runnning a loop t times to obtain actions
    for i in range(t):
        
        #Selecting a random action from the list of actions
        action = random.choice(actions)
        
        #Implementing the action according to the random input and changing values of reward depending upon the action
        if(action == "up"):
            if(y == 0):
                sum-=10
            else:
                y-=1
        elif(action == 'down'):
            if(y == n-1):
                sum-=10
            else:
                y+=1
        elif(action == 'left'):
            if(x == 0):
                sum-=10
            else:
                x-=1
        elif(action == 'right'):
            if(x == n-1):
                sum-=10
            else:
                x+=1
        elif(action == 'pickDirt'):
            if(room[x][y] != 0):
                sum += room[x][y]
                room[x][y] = 0
            else:
                sum -= 1
            
        #Appending the value of rewards in list to show at each time t
        reward.append(sum)
        
    #Returning the list
    return reward

#Taking input of the dimensions of the matrix
m = int(input("Enter dimension of row :"))
n = int(input("Enter dimension of column :"))

#Creating a 2D Matrix with all zeros
room = np.zeros((m, n))

#Adding 10 dirt particles in the matrix
add_dirt(room)
print(room)

#Taking input of time to print the reward at any time t
t=int(input("Enter the ammount of time you want to see :"))
reward = agents_Action(room, t)
for i in range(len(reward)):
    print(f"Reward at t = {i+1} is {reward[i]}")
```
