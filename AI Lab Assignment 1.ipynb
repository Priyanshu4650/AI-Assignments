# Monty Hall Game

You are on a game show and given the choice of picking one of the three boxes ($X$, $Y$, $Z$). Two of the boxes are empty and one contains a gift. You pick a box, say $X$, and the host (who knows what is inside each of the boxes) opens another box, say $Y$, which is empty. You now have the option of retaining your choice ($X$) or switching it (to $Z$). Is it to your advantage to switch your choice in order to get the gift? This advantage is defined as the fraction of times switching the choice leads to a win when the game is played 1000 times.

```python
# Monty Hall Game

# Importing the random library function
import random

# Defined a variable to count number of times we will win
count = 0

#n=int(input("Enter the number of times to play :"))

# Running a loop for 1000 times as said in the question
for i in range(1000):

    # The box selected by the player
    option = random.randint(1, 3)

    # The box contaning the gift
    prize = random.randint(1, 3)

    # The box which the host will remove
    host = random.randint(1, 3)

    # Making a condition so that the box that will be removed is not having gift and not chosen by the player
    while(host == prize and host == option):
        host = random.randint(1, 3)

    # Condition counting for switching
    if(option != prize):
        count += 1
    else:
        continue

# The probability of times it should switch
print(count/1000)

# condition to switch the box
if(count > 500):
    print("Switching is an option")

# Condition to not to switch the box
else:
    print("Don't switch")
```

# Game of Dice

    * If the sum is 2, 3 or 12 on the first roll, you lose. 
    * If the sum is 4, 5, 6, 8, 9 or 10 on the first roll, that sum becomes your point. 
2. Continue rolling the dice until the sum is 7 (game lost) or equal to your point (game won). 

If $n$ games are played and $q$ of these games are won, then the chance of winning is $q/n$. Sample simulation of the game for $n=10$ is as follows. What is the chance of winning at this game?

```python
# Game of Dice

# importing random library
import random

# Function to define win and loss after the first roll


def gameDice(n):

    # Running a loop until it returns a value
    while(True):

        # Getting both the random integers present on the dice
        a = random.randint(1, 6)
        b = random.randint(1, 6)

        print(f"Player rolled {a} + {b} ={a+b}")
        c = a+b

        # Condition to lose the game
        if(c == 7):
            print("Player loses")

            # returning 0 to end function and signify the loss
            return 0

        # Condition to win the game
        elif(c == n):
            print("Player wins")

            # Returning 1 to end the function and signify that it won
            return 1

# Function to decide for the first roll


def gameOfdice(n):

    # Win condition for the first throw
    if(n == 7 or n == 11):
        print("Player wins")
        # Ending the program signifying that it won
        return 1

    # Losing condition for the first throw
    elif(n == 2 or n == 3 or n == 12):
        print("Player loses")
        # Ending program with returning 0 signifying loss
        return 0

    # For other conditions returning the values obtaining from gameDice function
    return gameDice(n)

# Main function


def main():

    # Taking an input that how many games to be played
    n = int(input("Enter number of games to be played :"))

    # A variable to count number of wins
    wins = 0

    # Checking the output of the game n number of times
    for i in range(n):

        # GEtting the two random integers
        n1 = random.randint(1, 6)
        n2 = random.randint(1, 6)

        print(f"Player rolled {n1} + {n2} = {n1+n2}")

        # Total points scored
        n = n1+n2

        print("Point is ", n)

        # Output of to see win or loss
        a = gameOfdice(n)

        # Condition to increment the number of wins
        if(a == 1):
            wins += 1
        print("---")


# Printing the number of wins by calling out the main function
print("Total wins ", main())
```

# Central Limit Theorem

Repeat this $10^5$ times and plot the frequency of the average values (i.e., how many times each average value occurs)  as a histogram. Repeat this experiment with $n \in \{5,10,30,100,1000\}$.
